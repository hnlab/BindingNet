```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
models = ['PDBbind_minimized', 'PDBbind_minimized_intersected_Uw', 'PLANet_Uw', 'PDBbind_minimized_intersected_Uw_union_Uw', 'PDBbind_minimized_union_Uw', 'PLANet_all']
```

```python
model = models[0]
i = 1
```

```python
test_res = pd.read_csv(f'/pubhome/xli02/project/PLIM/deep_learning/RFscore/test_res/{model}_Rm_core/{i}/final_result_VR1_RF.csv')
test_res['training_set'] = model
test_res['repeat'] = i
test_res
```

```python
res_df_list = []
for model in models:
    for i in range(5):
        test_res = pd.read_csv(f'/pubhome/xli02/project/PLIM/deep_learning/RFscore/test_res/{model}_Rm_core/{i+1}/final_result_VR1_RF.csv')
        test_res['training_set'] = model
        test_res['repeat'] = i+1
        res_df_list.append(test_res)
```

```python
len(res_df_list)
```

```python
all_res_df = pd.concat(res_df_list)
all_res_df
```

```python
all_res_df.groupby(['name', 'training_set']).mean()
```

```python
all_res_df.groupby(['name', 'training_set']).median()
```

```python
grouped_median = all_res_df.groupby(['name', 'training_set']).median()[['r2', 'Rp', 'mae']].reset_index()
grouped_median
```

```python
grouped = all_res_df.groupby('name')
test_type = 'Core'
metric = 'Rp'
```

```python
test_df = grouped.get_group(test_type)
fig, ax= plt.subplots()
sns.boxplot(x="training_set", y=metric, data=test_df, order = models, linewidth=2.5)
sns.swarmplot(x="training_set", y=metric, data=test_df, order = models, size = 4, edgecolor="black", linewidth=0.7)
ax.set_title(f'{metric} on {test_type} set')
fig.autofmt_xdate()
vertical_offset = test_df[metric].median() * 0.01 # offset from median for display
for i, modl in enumerate(models):
    median_metric = round(grouped_median.loc[(grouped_median['training_set'] == modl) & (grouped_median['name'] == test_type)][metric].values[0], 3)
    ax.text(i, median_metric-vertical_offset, median_metric, horizontalalignment='center',size='small', weight='semibold')
```

```python
grouped = all_res_df.groupby('name')
test_type = 'Core'
metric = 'mae'
test_df = grouped.get_group(test_type)
fig, ax= plt.subplots()
sns.boxplot(x="training_set", y=metric, data=test_df, order = models, linewidth=2.5)
sns.swarmplot(x="training_set", y=metric, data=test_df, order = models, size = 4, edgecolor="black", linewidth=0.7)
ax.set_title(f'{metric} on {test_type} set')
fig.autofmt_xdate()
vertical_offset = test_df[metric].median() * 0.01 # offset from median for display
for i, modl in enumerate(models):
    median_metric = round(grouped_median.loc[(grouped_median['training_set'] == modl) & (grouped_median['name'] == test_type)][metric].values[0], 3)
    ax.text(i, median_metric-vertical_offset, median_metric, horizontalalignment='center',size='small', weight='semibold')
```

```python
test_types = ['Train', 'Valid', 'Self_test', 'Core', 'Core_intersected_Uw']
```

```python
grouped = all_res_df.groupby('name')
for metric in ['r2', 'Rp', 'mae']:
    out_dir = f'/pubhome/xli02/project/PLIM/deep_learning/RFscore/4-evaluation/res/{metric}'
    if not Path(out_dir).exists():
        Path(out_dir).mkdir()
    for test_type in test_types:
        test_df = grouped.get_group(test_type)
        fig, ax= plt.subplots()
        sns.boxplot(x="training_set", y=metric, data=test_df, order = models, linewidth=2.5)
        sns.swarmplot(x="training_set", y=metric, data=test_df, order = models, size = 4, edgecolor="black", linewidth=0.7)
        ax.set_title(f'{metric} on {test_type} set')
        fig.autofmt_xdate()
        vertical_offset = test_df[metric].median() * 0.01 # offset from median for display
        for i, modl in enumerate(models):
            median_metric = round(grouped_median.loc[(grouped_median['training_set'] == modl) & (grouped_median['name'] == test_type)][metric].values[0], 3)
            ax.text(i, median_metric-vertical_offset, median_metric, horizontalalignment='center',size='small', weight='semibold')
        if metric in ['r2', 'Rp']:
            ax.set_ylim(min(test_df[metric])-0.02, min(max(test_df[metric])+0.02, 1))
        plt.savefig(f'{out_dir}/{metric}_on_{test_type}_set.png', dpi=300, bbox_inches='tight')
        plt.close()
```
